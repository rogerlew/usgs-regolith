```python
#stdlib imports
import subprocess
from time import sleep
import os

#third party imports
import numpy as np
```

```python
def write_files(filename, lines):
    with open(filename, 'w+') as f:
        f.writelines(lines)
```

```python
##  REGOLITH  ##

## File naming
filename = "rg_in.txt" 
dir1 = os.path.abspath('.')
sys_name = os.name
if sys_name == 'nt':
    cmd = os.path.abspath('./regolith.exe')
else:
    cmd = os.path.abspath('./regolith')
print('Executable path name:', cmd)

## Iterative process 
## User Inputs
model = 'LCSD'
#sc = np.arange(30, 60, 5) #(low, high, interval)
dif_ratio = np.arange(0.01, 0.03, 0.01)
C1 = np.arange(0.01, 0.02, 0.01) 
C0 = np.arange(0.25, 2.25, 0.25)
C2 = np.arange(1, 2.5, 0.5)
power = np.arange(1, 1.75, 0.25)

if (model == 'DRS1'): 
    for d in C1:
        for c in C0:
            f = open(filename)
            lines = f.readlines()
            f.close()
            
            # rewriting variables
            variable_line = lines[10].split(',')
            variable_line[2] = d
            variable_line[5] = c
            variable_line = ",".join([str(i) for i in variable_line])
            variable_fname = "%s_%s\n" % (d, c)
            
            # replacing lines
            lines[10] = variable_line 
            lines[3] = model + "\n"
            lines[28] = variable_fname
            fname1 = os.path.join(dir1, 'rg_in.txt')
            write_files(fname1, lines)
            print("C1, C0 = ", d, c)
            subprocess.run(cmd, cwd = dir1)
            sleep(10)
    print("Finished!")
            
elif (model == 'DRS2'): 
    for d in C1:
        for c in C0:
            for b in power:
                f = open(filename)
                lines = f.readlines()
                f.close()

                # rewriting variables
                variable_line = lines[10].split(',')
                variable_line[2] = d
                variable_line[5] = c
                variable_line1 = lines[12]
                variable_line1 = str(b) 
                variable_line = ",".join([str(i) for i in variable_line])
                variable_fname = "%s_%s_%s\n" % (d, c, b)

                # replacing lines
                lines[10] = variable_line
                lines[12] = variable_line1 + "\n"
                lines[3] = model + "\n"
                lines[28] = variable_fname
                fname1 = os.path.join(dir1, 'rg_in.txt')
                write_files(fname1, lines)
                print("C1, C0, power = ", d, c, b)
                subprocess.run(cmd, cwd = dir1)
                sleep(10)
    print("Finished!")
                
elif (model == 'DRS3'): 
    for d in C1:
        for c in C0:
            for e in C2:
                f = open(filename)
                lines = f.readlines()
                f.close()

                # rewriting variables
                variable_line = lines[10].split(',')
                variable_line[2] = d
                variable_line[5] = c
                variable_line[6] = e
                variable_line = ",".join([str(i) for i in variable_line])
                variable_fname = "%s_%s_%s\n" % (d, c, e)

                # replacing lines
                lines[10] = variable_line
                lines[3] = model + "\n"
                lines[28] = variable_fname
                fname = os.path.join(dir1, 'rg_in.txt')
                write_files(fname, lines)
                print("C1, C0, C2 = ", d, c, e)
                subprocess.run(cmd, cwd = dir1)
                sleep(10)    
    print("Finished!")
            
elif (model == 'WNDX'):
    for c in C0:
        for b in power:
            f = open(filename)
            lines = f.readlines()
            f.close()

            # rewriting variables
            variable_line = lines[10].split(',')
            variable_line[5] = c
            variable_line1 = lines[12]
            variable_line1 = str(b)
            variable_line = ",".join([str(i) for i in variable_line])
            variable_fname = "%s_%s\n" % (c, b)

            # replacing lines
            lines[10] = variable_line 
            lines[12] = variable_line1 + "\n"
            lines[3] = model + "\n"
            lines[28] = variable_fname
            fname = os.path.join(dir1, 'rg_in.txt')
            write_files(fname, lines)
            print("C0, power = ", c, b)
            subprocess.run(cmd, cwd = dir1)
            sleep(10)
    print("Finished!")
           
elif (model == 'NSDA' or model == 'NASD'):
    for d in dif_ratio:
        for b in power:
            f = open(filename)
            lines = f.readlines()
            f.close()

            # rewriting variables
            variable_line = lines[10].split(',')
            variable_line[2] = d
            variable_line1 = lines[12]
            variable_line1 = str(b)
            variable_line = ",".join([str(i) for i in variable_line])
            variable_fname = "%s_%s\n" % (d, b)
            
            # replacing lines
            lines[10] = variable_line 
            lines[12] = variable_line1 + "\n"
            lines[3] = model + "\n"
            lines[28] = variable_fname
            fname = os.path.join(dir1, 'rg_in.txt')
            write_files(fname, lines)
            print("dif_ratio, power = ", d, b)
            subprocess.run(cmd, cwd = dir1)
            sleep(10)
    print("Finished!")
            
elif (model == 'LCSD' or model == 'NSD' or model == 'NDSD'):
    for d in dif_ratio:
        f = open(filename)
        lines = f.readlines()
        f.close()
        
        # rewriting variables
        variable_line = lines[10].split(',')
        variable_line[2] = d
        variable_line = ",".join([str(i) for i in variable_line])
        variable_fname = "dif_%s\n" % (d)
        
        # replacing lines
        lines[10] = variable_line
        lines[3] = model + "\n"
        lines[28] = variable_fname
        fname = os.path.join(dir1, 'rg_in.txt')
        write_files(fname, lines)
        print("dif_ratio = ", d)
        subprocess.run(cmd, cwd = dir1)
        if (lines[3] == 'NDSD'):
            sleep(90)
        else:
            sleep(10)
    print("Finished!")
else:
    print('Invalid model type, Correct model name and try again.')
```
